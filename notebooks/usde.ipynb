```python
query = """
SELECT * FROM llama.circulating
WHERE id = 146
"""
df= pd.read_sql(query, engine)

l = df['time'].max()

tvl = df[(df['time'] == l)]['circulating'].sum()

n_chains = df[
    (df['time'] == l) 
]['chain'].nunique()


df = df.groupby(['time', "chain"]).agg(
    tvl=('circulating', 'sum')
).reset_index()
df

fig = px.bar(df, x="time", y="tvl",  title=f"Current TVL: {tvl/1e9:.2f}B")
fig.update_layout(
    xaxis_title="Date",
    yaxis_title="TVL (USD)",
    legend_title_text="Stablecoin"
)
fig.show()
```
